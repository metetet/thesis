```python
import matplotlib.pyplot as plt
import numpy as np

results_dict = {
    "FSL (Art To Faces)": 0.7755102040816326,
    "FSL (Faces To Art)": 0.7346938775510204,
    "Fine-Tuned (Mixed)": 0.782608695652174,
    "Fine-Tuned (Art)": 0.9056603773584906,
    "Fine-Tuned (Faces)": 0.0,
    "Original (Organika)": 0.7450980392156863,   
}

results_faces = {
    "FSL (Art To Faces)": 0.672566371681416,
    "FSL (Faces To Art)": 0.9304245283018868,
    "Fine-Tuned (Mixed)": 0.944206008583691,
    "Fine-Tuned (Art)": 0.3943139678615575,
    "Fine-Tuned (Faces)": 0.9849435382685069,
    "Original (Organika)": 0.669,   
}

results_art = {
    "FSL (Art To Faces)": 0.9315589353612167,
    "FSL (Faces To Art)": 0.8063194850789935,
    "Fine-Tuned (Mixed)": 0.9599012954966071,
    "Fine-Tuned (Art)": 0.9651376146788991,
    "Fine-Tuned (Faces)": 0.023492560689115115,
    "Original (Organika)": 0.816,   
}

labels = list(results_dict.keys())
dogs = list(results_dict.values())
faces = list(results_faces.values())
art = list(results_art.values())

y = np.arange(len(labels))  
height = 0.25

plt.figure(figsize=(8, 6))
bars_faces = plt.barh(y + height, faces, height, label='Faces')
bars_art = plt.barh(y, art, height, label='Art')
bars_dogs = plt.barh(y - height, dogs, height, label='Dogs')
plt.xlabel('F1 Score') 
plt.ylabel('Model')
plt.title('F1 Scores for Different Models on Faces, Art, and Dogs Test Splits')
plt.yticks(y, labels, rotation=35)
plt.legend(['Faces', 'Art', 'Dogs'])

for bars in [bars_dogs, bars_faces, bars_art]:
    for bar in bars:
        plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2,
                 f'{bar.get_width():.3f}', va='center')

plt.xlim(0, 1.1) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = list(results_dict.keys())
dogs = np.array(list(results_dict.values()))
faces = np.array(list(results_faces.values()))
art = np.array(list(results_art.values()))

y = np.arange(len(models))
height = 0.8  # full bar height

plt.figure(figsize=(10, 6))

# Plot each domain as a bar at the same y position with transparency
bars_dogs = plt.barh(y, dogs, height, label='Dogs', color='tab:blue')
bars_faces = plt.barh(y, faces, height, label='Faces', color='tab:orange')
bars_art = plt.barh(y, art, height, label='Art', color='tab:green')

plt.xlabel('F1 Score')
plt.title('F1 Scores for Different Models (Overlayed by Domain)')
plt.yticks(y, models, rotation=35)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

# Add values
for bars in [bars_dogs, bars_faces, bars_art]:
    for bar in bars:
        plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2,
                 f'{bar.get_width():.3f}', va='center')

plt.xlim(0, 1.1)
plt.tight_layout()
plt.show()
```
