```python
import matplotlib.pyplot as plt
import numpy as np

results_dict = {
    "FSL (Art To Faces)": 0.7755102040816326,
    "FSL (Faces To Art)": 0.7346938775510204,
    "Fine-Tuned (Mixed)": 0.782608695652174,
    "Fine-Tuned (Art)": 0.9056603773584906,
    "Fine-Tuned (Faces)": 0.0,
    "Original (Organika)": 0.7450980392156863,   
}

plt.figure(figsize=(7, 6))
bars = plt.barh(list(results_dict.keys()), list(results_dict.values()))
plt.xlabel('F1 Score') 
plt.ylabel('Model')
plt.title('F1 Scores for Different Models on The Dogs Test Split')
plt.yticks(rotation=35)

for bar in bars:
    if bar == bars[4]:
        plt.text(bar.get_width() + 0.01,  
                bar.get_y() + bar.get_height() / 2, 
                f'{bar.get_width():.3f} (Accuracy: 0.44)',            
                va='center')             
    else:
        plt.text(bar.get_width() + 0.01,   
                bar.get_y() + bar.get_height() / 2,  
                f'{bar.get_width():.3f}',     
                va='center')

plt.xlim(0, 1.1) 
plt.show()
```
