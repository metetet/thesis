```python
import matplotlib.pyplot as plt
import numpy as np

# F1 scores for different models on Faces dataset
f1_faces_original = 0.6688
f1_faces_finetuned_faces = 0.9912
f1_faces_finetuned_art = 0.2641
f1_faces_finetuned_both = 0.7  # Assumed value

# F1 scores for different models on Art dataset
f1_art_original = 0.8159
f1_art_finetuned_faces = 0.0356
f1_art_finetuned_art = 0.9564
f1_art_finetuned_both = 0.7  # Assumed value

# Labels and data
models = ['Original Model', 'Fine-Tuned (Faces)', 'Fine-Tuned (Art)', 'Fine-Tuned (Mix)']
f1_faces_scores = [f1_faces_original, f1_faces_finetuned_faces, f1_faces_finetuned_art, f1_faces_finetuned_both]
f1_art_scores = [f1_art_original, f1_art_finetuned_faces, f1_art_finetuned_art, f1_art_finetuned_both]

x = np.arange(len(models))  # Bar positions

# Plot for Faces dataset
plt.figure(figsize=(8, 6))
plt.bar(x, f1_faces_scores, color=['blue', 'orange', 'green', 'red'])
plt.xticks(x, models, rotation=20)
plt.ylabel("F1 Score")
plt.ylim(0, 1)
plt.title("F1 Scores on Faces Dataset")

# Add labels on bars
for i, v in enumerate(f1_faces_scores):
    if i != 1:
        plt.text(i, v + 0.01, f"{v:.4f}", ha='center')
    else:
        plt.text(i, v - 0.024, f"{v:.4f}", ha='center')
    

plt.show()

# Plot for Art dataset
plt.figure(figsize=(8, 6))
plt.bar(x, f1_art_scores, color=['blue', 'orange', 'green', 'red'])
plt.xticks(x, models, rotation=20)
plt.ylabel("F1 Score")
plt.ylim(0, 1)
plt.title("F1 Scores on Art Dataset")

# Add labels on bars
for i, v in enumerate(f1_art_scores):
    plt.text(i, v + 0.01, f"{v:.4f}", ha='center')

plt.show()
```
